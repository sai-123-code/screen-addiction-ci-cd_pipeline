```python
# âœ… Import Required Libraries
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# âœ… Load Dataset
df = pd.read_csv('teen_phone_addiction_dataset.csv')
df=df[['Age', 'Gender',
       'Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance',
       'Social_Interactions', 'Exercise_Hours', 'Anxiety_Level',
       'Depression_Level', 'Self_Esteem', 'Parental_Control',
       'Screen_Time_Before_Bed', 'Phone_Checks_Per_Day', 'Apps_Used_Daily',
       'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education',
       'Phone_Usage_Purpose', 'Family_Communication', 'Weekend_Usage_Hours',
       'Addiction_Level']]
df.dropna(inplace=True)

# âœ… Binning Addiction_Level into 3 Categories
df['Addiction_Level_Category'] = pd.cut(
    df['Addiction_Level'],
    bins=[-np.inf, 3.5, 7.5, np.inf],
    labels=[0, 1, 2]  # 0 = Low, 1 = Medium, 2 = High
)

# Replace original column with categorical version
df.drop(columns=['Addiction_Level'], inplace=True)
df.rename(columns={'Addiction_Level_Category': 'Addiction_Level'}, inplace=True)

# âœ… Remove rare classes (<2 samples)
class_counts = df['Addiction_Level'].value_counts()
valid_classes = class_counts[class_counts >= 2].index
df = df[df['Addiction_Level'].isin(valid_classes)]

# âœ… Encode categorical columns
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# âœ… Define features and target
X = df.drop('Addiction_Level', axis=1)
y = df['Addiction_Level'].astype(int)

# âœ… Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# âœ… Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)




```

```python
# âœ… Train Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# âœ… Evaluate
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"âœ… Accuracy: {acc:.4f}")
print("ðŸ“Š Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("ðŸ“„ Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# âœ… Save model and scaler as pickle
with open("logistic_model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# âœ… Save label encoders (if needed for Flask inference)
with open("label_encoders.pkl", "wb") as f:
    pickle.dump(label_encoders, f)
```
